```python
# HILL CIPHER ENCRYPTION
import numpy as np
#alphabets
alp="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
msg="ACT"
n=3
key="GYBNQKURP"# 3*3
keymat=[alp.index(i) for i in key] # get index from above alp string
keymat=np.array(keymat).reshape((n,n)) # convert to numpy matrox and reshaping to n*n
print("key matrix",keymat,sep="\n")
msgmat=[alp.index(i) for i in msg]
print(len(msgmat))
msgmat=np.array(msgmat).reshape((n,-1))
print("msg matrix",msgmat,sep="\n")
ciphermat=keymat.dot(msgmat)
ciphermat=ciphermat%26
print("cipher matrix",ciphermat,sep="\n")
ciphermat=list(np.squeeze(ciphermat))
ciphertext="".join(alp[i] for i in ciphermat)
ciphertext
```
